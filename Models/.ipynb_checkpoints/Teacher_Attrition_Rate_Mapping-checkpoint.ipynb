```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_counties_in_North_Carolina'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
table = soup.findAll('table')[1]
table
```

```python
headers =[]
header = []
# header is in the first row
for val in table.find_all('tr')[0].find_all('th'):
    #print(val.text)
    header.append(val.text.replace('\n',''))
headers.append(header)
#headers
```

```python
countyFP =pd.DataFrame([])
countyFP =pd.DataFrame(columns = list(headers)[0])

for row in table.find_all('tr')[1:]:
    data =[]
    #print(row)
    for val in row.find_all('th'):
        fieldVal = val.text.replace('\n','')
        data.append(fieldVal)
    for val in row.find_all('td'):
        fieldVal = val.text.replace('\n','')
        data.append(fieldVal)
    #print(data)
    countyFP.loc[len(countyFP)] = data

countyFP.columns = ['County','FIPS_Code','County_Seat'
                    ,'Created','Origin','Etymoloty'
                    ,'Population_2017','Area'
                    ,'Map']
```

```python
schoolYear = 2017
dataDir = '../' + str(schoolYear) + '/School Datasets/'
schoolData = pd.read_csv(dataDir + 'PublicSchools2017.csv', low_memory=False, dtype={'unit_code': object}) 
```

```python
teacherbyCounty = schoolData[['County','_1yr_tchr_trnovr_pct']]


teacherbyCounty = teacherbyCounty.groupby(['County'], as_index=False).mean()
```

```python
#Merge accDrillDownFemale data
teacherMap = countyFP.merge(teacherbyCounty,how='left',on='County', suffixes=('', '_Drop'))

print('*********************************After: teacherbyCounty Data**********************')
teacherMap.info(verbose=False)
```

```python
teacherMap.to_csv("2017_Teach_Attrition_Map.csv", sep=',', index=False)
```
