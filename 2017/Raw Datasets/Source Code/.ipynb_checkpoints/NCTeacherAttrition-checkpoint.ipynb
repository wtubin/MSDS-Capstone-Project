```python
pfr = PyPDF2.PdfFileReader(open(PDFfilename,"rb"))
```

```python
NewPDFfilename = "TeacherAttrition_ByReasonsCategories.pdf"
writer = PyPDF2.PdfFileWriter()
    
for i in range(27,30):
    pg = pfr.getPage(i-1)
    writer.addPage(pg)
with open(NewPDFfilename,"wb") as outputStream:
    writer.write(outputStream)
```

```python
import re
def count_pdf_pages(file_path):
    rxcountpages = re.compile("/Type\s*/Page([^s]|$)", re.MULTILINE|re.DOTALL)
    with open(file_path, "rb") as temp_file:
        return len(rxcountpages.findall(temp_file.read()))
```
